```python
import pygame
import random

print("Let's play TRAFFIC TESTING!")

# Initialize Pygame
pygame.init()

# Constants
WINDOW_WIDTH = 600
WINDOW_HEIGHT = 400
CAR_WIDTH = 50
CAR_HEIGHT = 100

# Colors
WHITE = (255, 255, 255) # colors to use later i the app
BLACK = (0, 0, 0)

# Create window
window = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption('Traffic Testing Game') 

clock = pygame.time.Clock()

# Car Blue Rectangle
car_img = pygame.Surface((CAR_WIDTH, CAR_HEIGHT), pygame.SRCALPHA)
pygame.draw.rect(car_img, (0, 100, 200), [0, 0, CAR_WIDTH, CAR_HEIGHT])

# Obstacle variables
obstacles = []
obstacle_speed = 5

# Car starting position and speed
car_x = WINDOW_WIDTH // 2 - CAR_WIDTH // 2
car_y = WINDOW_HEIGHT - CAR_HEIGHT - 20
car_speed = 5

# Main game loop
running = True

while running:
    window.fill(WHITE)  # make background white ising colors form above

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False  # End the game if the window is closed

    # Control the car movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and car_x > 0:
        car_x -= car_speed
    elif keys[pygame.K_RIGHT] and car_x < WINDOW_WIDTH - CAR_WIDTH:
        car_x += car_speed
    if keys[pygame.K_UP] or keys[pygame.K_w]:
        car_y -= car_speed
    if keys[pygame.K_DOWN] or keys[pygame.K_s]:
        car_y += car_speed

    # Create obstacles randomly
    if random.randint(1, 75) == 1:  # Obstacle randomness
        obstacle_width = random.randint(50, 150)
        obstacle_x = random.randint(0, WINDOW_WIDTH - obstacle_width)
        obstacle_y = -20
        obstacles.append(pygame.Rect(obstacle_x, obstacle_y, obstacle_width, 20))  # Create a rectangle for road blocks

    # Manage and update obstacle movements
    for obstacle in obstacles:
        pygame.draw.rect(window, BLACK, obstacle)  # Draw the obstacles
        obstacle.y += obstacle_speed  # Update the obstacle's position

        # Check for collision between car and obstacles
        if obstacle.colliderect(pygame.Rect(car_x, car_y, CAR_WIDTH, CAR_HEIGHT)):
            print("Collision Detected!")
            running = False  # Stop the game on collision

        if obstacle.y > WINDOW_HEIGHT:
            obstacles.remove(obstacle)  # Remove obstacles that are out of the window

    window.blit(car_img, (car_x, car_y))  # Show the "car"

    pygame.display.update()  # Update the window display
    clock.tick(60)  # Control frame rate

pygame.quit()  # Quit Pygame after the game ends

```
